```python
import dateutil.parser
start = dateutil.parser.parse("2017-09-01T00:00:00Z")  
end   = dateutil.parser.parse("2017-09-30T00:00:00Z")
```

```python
from pymongo import MongoClient
client = MongoClient("mongodb://127.0.0.1:27017")
db = client["<db name>"]

collection = db["<collection name>"]
contents = collection.find({"datetime" : {"$gte":start,"$lte":end}},{"content":1,"_id":0})

liberty_news_cleaned_lsit=[]
for content in contents:
    liberty_news_cleaned_lsit.extend(content["content"].split(" "))
```

```python
from collections import Counter
wordcount = Counter(liberty_news_cleaned_lsit)
for i in wordcount.most_common(1000):
    print(i)
```
