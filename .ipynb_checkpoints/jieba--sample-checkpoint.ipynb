```python
import jieba
import re
import copy
from datetime import datetime
from pymongo import MongoClient
import pprint
```

```python
### 載入停用字檔案，並做成一個list
stopwords_list = [line.strip() for line in open('stopwords.txt',"r").readlines()]

### 載入同義字檔案，並做成一個字典
syn_dict = {}
with open("syn.txt","r") as f :
    for line in f:
        for word in line.strip("\n").split("\t")[1:]:
            syn_dict[word] = line.strip("\n").split("\t")[0]

### 載入中文斷詞字典
jieba.set_dictionary("dict.txt")

### 載入MongoDB裡面的新聞
client = MongoClient("mongodb://10.120.37.108:27017")
db = client["news"]
collection = db["united_daily_news"]        #<<<<<<<<<<<<<<<<<<<<================================= 輸入自己的新聞名子
news_list_temp = collection.find({}).limit(50) #<<<<<<<<<<<<<<<<<================================= 改輸出筆數
news_list = list(news_list_temp)
```

```python
def stopwords(w):
    if w not in stopwords_list:
        return w

def syn(w):
    if w in syn_dict.keys():
        w=syn_dict[w]
        return w
    else:
        return w

def cut(news):
    w = jieba.cut(news, cut_all=False ,HMM=True)
    return w

def regular(w):
    line = re.findall('[a-z|A-Z|\u4e00-\u9fa5]+', w)
    if len(line) > 0:
        return line
```

```python
def text_cleaning(paragraph):
    me_words = []
    words = cut(paragraph) 
    for w in words:
        w = regular(w)
        if w is not None:
            w_stopwords = stopwords(w[0])
            if w_stopwords is not None:
                w_syn = syn(w_stopwords)
                me_words.append(w_syn)
    return " ".join(me_words)
```

```python
def cleaned_news(news_list):
    cleaned_news_list = []
    for news in copy.deepcopy(news_list):
        news["datetime"]= datetime.strptime(news["datetime"],"%Y-%m-%d")
        news["content"] = text_cleaning(news["content"])
        cleaned_news_list.append(news)
    return cleaned_news_list
```

```python
def main():
    db_cleaned = client["news_cleaned"]
    collection = db_cleaned["united_daily_news_cleaned"]  #<<<<<<<<<<<<<<<<<<<<============================= 輸入自己的新聞名子
    pprint.pprint(cleaned_news(news_list))                
    
    #     collection.insert_many(cleaned_news(news_list)) #<<<<<<<<<<<<<<<<<<<============================= 修改好就存MongoDB
```
